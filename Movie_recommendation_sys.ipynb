```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
#Merging the two datasets
movies = movies.merge(credits,on='title')
movies.shape
```

```python
credits.shape
```

```python
#important features column
#genres
#id
#keywords
#title
#overview
#cast
#crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
#dropping missing values from movies

movies.dropna(inplace=True)
```

```python
import ast

def convert(obj):
  L=[]
  for i in ast.literal_eval(obj):    #to convert string of lists to list (ast.literal_eval()) if we without this function it will show error
    L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```
